```python
import matplotlib.pyplot as plt
import numpy as np

# Datos proporcionados
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
ventas_telefonos = [120, 110, 130, 140, 150, 160, 155, 150, 145, 140, 135, 130]
ventas_laptops = [80, 85, 90, 95, 100, 110, 105, 100, 95, 90, 85, 80]
ventas_tabletas = [50, 60, 65, 70, 75, 80, 85, 90, 95, 100, 105, 110]

# Ejercicio 1: Anlisis de Tendencias con Subplots
plt.figure(figsize=(12, 8))

# Subplot para telfonos
plt.subplot(3, 1, 1)
plt.plot(meses, ventas_telefonos, label='Telfonos', color='blue')
plt.title('Ventas Mensuales de Telfonos')
plt.xlabel('Mes')
plt.ylabel('Ventas (unidades)')
plt.legend()

# Subplot para laptops
plt.subplot(3, 1, 2)
plt.plot(meses, ventas_laptops, label='Laptops', color='green')
plt.title('Ventas Mensuales de Laptops')
plt.xlabel('Mes')
plt.ylabel('Ventas (unidades)')
plt.legend()

# Subplot para tabletas
plt.subplot(3, 1, 3)
plt.plot(meses, ventas_tabletas, label='Tabletas', color='orange')
plt.title('Ventas Mensuales de Tabletas')
plt.xlabel('Mes')
plt.ylabel('Ventas (unidades)')
plt.legend()

plt.tight_layout()
plt.show()

# Ejercicio 2: Visualizacin de Proporciones con Grfico de Torta
ventas_anuales = np.array([sum(ventas_telefonos), sum(ventas_laptops), sum(ventas_tabletas)])
categorias = ['Telfonos', 'Laptops', 'Tabletas']

plt.figure(figsize=(8, 8))
plt.pie(ventas_anuales, labels=categorias, autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'lightcoral'])
plt.title('Proporcin de Ventas Anuales por Categora')
plt.show()

# Ejercicio 3: Anlisis de Variabilidad con Boxplots
plt.figure(figsize=(10, 6))

# Boxplot para telfonos
plt.subplot(1, 3, 1)
plt.boxplot(ventas_telefonos)
plt.title('Variabilidad de Ventas de Telfonos')

# Boxplot para laptops
plt.subplot(1, 3, 2)
plt.boxplot(ventas_laptops)
plt.title('Variabilidad de Ventas de Laptops')

# Boxplot para tabletas
plt.subplot(1, 3, 3)
plt.boxplot(ventas_tabletas)
plt.title('Variabilidad de Ventas de Tabletas')

plt.tight_layout()
plt.show()
```
